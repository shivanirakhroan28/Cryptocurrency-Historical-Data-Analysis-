```python
bitcoin.columns= bitcoin.columns.map(lambda x: str(x)+'_bit')
ethereum.columns= ethereum.columns.map(lambda x: str(x)+'_eth')
litecoin.columns= litecoin.columns.map(lambda x: str(x)+'_lit')
monero.columns= monero.columns.map(lambda x: str(x)+'_mon')
neo.columns= neo.columns.map(lambda x: str(x)+'_neo')
qtum.columns= qtum.columns.map(lambda x: str(x)+'_qtum')
ripple.columns= ripple.columns.map(lambda x: str(x)+'_ripple')
```

```python
m1= pd.merge(bitcoin,ethereum,how="inner",left_on="Date_bit",right_on="Date_eth")
m2= pd.merge(m1,litecoin,how="inner",left_on="Date_bit",right_on="Date_lit")
m3= pd.merge(m2,monero,how="inner",left_on="Date_bit",right_on="Date_mon")
m4= pd.merge(m3,neo,how="inner",left_on="Date_bit",right_on="Date_neo")
m5= pd.merge(m4,qtum,how="inner",left_on="Date_bit",right_on="Date_qtum")
final= pd.merge(m5,ripple,how="inner",left_on="Date_bit",right_on="Date_ripple")
```
